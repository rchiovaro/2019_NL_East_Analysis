# Exploring the NL East with pybaseball

A brief analysis and visualization of the NL East division in 2019. Teams: Mets, Braves, Nationals, Phillies, and Marlins.

```python
import pybaseball as pyball
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Get all NL East Team Records for 2019
year = 2019
mets = pyball.schedule_and_record(year, 'NYM')
braves = pyball.schedule_and_record(year, 'ATL')
nationals = pyball.schedule_and_record(year, 'WSN')
phillies = pyball.schedule_and_record(year, 'PHI')
marlins = pyball.schedule_and_record(year, 'MIA')
```

## Summary Statistics
A quick glance at the summary statistics for each team in order of their ending division rank.

Looking at these statistics, you can see the similarites and differences in overall performance. Mean Runs are very similar for the Braves and the Nationals, who both slightly edge out the Mets and Phillies, with the Marlins in 5th. However, in Runs Allowed, the Mets, Braves, and Nationals are all very similar with a slight edge to the Nationals, and the Phillies and Marlins lagging behind .

Interestingly, the highest average attendance belongs to the Phillies, who ended the season ranked 4th in the division.

## A look at Wins, Win Percentages, and Games Back
Comparing the wins, win percentages, and games back for each team of the NL East

First we gather each team's Wins, Losses, and Win Percentages at the time of each game

```python
# Gets wins-to-date, losses-to-date, and win_percent-to-date for each team
mets['Wins'] = np.where(mets['W/L']=='W', 1, (np.where(mets['W/L']=='W-wo', 1, 0))).cumsum()
mets['Losses'] = np.where(mets['W/L']=='L', 1, (np.where(mets['W/L']=='L-wo', 1, 0))).cumsum()
mets['Win_Percentage'] = mets['Wins'] / (mets['Wins'] + mets['Losses'])

braves['Wins'] = np.where(braves['W/L']=='W', 1, (np.where(braves['W/L']=='W-wo', 1, 0))).cumsum()
braves['Losses'] = np.where(braves['W/L']=='L', 1, (np.where(braves['W/L']=='L-wo', 1, 0))).cumsum()
braves['Win_Percentage'] = braves['Wins'] / (braves['Wins'] + braves['Losses'])

nationals['Wins'] = np.where(nationals['W/L']=='W', 1, (np.where(nationals['W/L']=='W-wo', 1, 0))).cumsum()
nationals['Losses'] = np.where(nationals['W/L']=='L', 1, (np.where(nationals['W/L']=='L-wo', 1, 0))).cumsum()
nationals['Win_Percentage'] = nationals['Wins'] / (nationals['Wins'] + nationals['Losses'])

phillies['Wins'] = np.where(phillies['W/L']=='W', 1, (np.where(phillies['W/L']=='W-wo', 1, 0))).cumsum()
phillies['Losses'] = np.where(phillies['W/L']=='L', 1, (np.where(phillies['W/L']=='L-wo', 1, 0))).cumsum()
phillies['Win_Percentage'] = phillies['Wins'] / (phillies['Wins'] + phillies['Losses'])

marlins['Wins'] = np.where(marlins['W/L']=='W', 1, (np.where(marlins['W/L']=='W-wo', 1, 0))).cumsum()
marlins['Losses'] = np.where(marlins['W/L']=='L', 1, (np.where(marlins['W/L']=='L-wo', 1, 0))).cumsum()
marlins['Win_Percentage'] = marlins['Wins'] / (marlins['Wins'] + marlins['Losses'])
```

### Graphing Wins and Win Percentages
Here we can see that the Braves take a lead in the first third of the season, which they would hold for the duration of the 2019 season. At about game 110 the Nationals start to come on strong and begin to threaten the Braves in the division. Despite the gap being narrowed significantly, the Braves were able to fend them off successfully, winning their second straight division crown.

```python
# Graph Wins Comparison #
plt.rcParams["figure.figsize"] = (8,6)

plt.plot(mets['Wins'], label='Mets', c='darkorange')
plt.plot(braves['Wins'], label='Braves', c='navy')
plt.plot(nationals['Wins'], label='Nationals', c='gray')
plt.plot(phillies['Wins'], label='Phillies', c='red')
plt.plot(marlins['Wins'], label='Marlins', c='aqua')

plt.xticks(np.arange(0, len(braves.index), step=10))
plt.xlabel('Game')
plt.ylabel('Wins')

plt.legend(loc='lower right')

plt.title("NL East Wins Comparison ({})".format(year))
```

```python
# Graph Win Percentage Comparison #
plt.rcParams["figure.figsize"] = (8,6)

plt.plot(mets['Win_Percentage'], label='Mets', c='darkorange')
plt.plot(braves['Win_Percentage'], label='Braves', c='navy')
plt.plot(nationals['Win_Percentage'], label='Nationals', c='gray')
plt.plot(phillies['Win_Percentage'], label='Phillies', c='red')
plt.plot(marlins['Win_Percentage'], label='Marlins', c='aqua')

plt.xticks(np.arange(0, len(braves.index), step=10))
plt.xlabel('Game')
plt.ylabel('Win Percentage')

plt.legend(loc='lower right')

plt.title("NL East Win Percentage Comparison ({})".format(year))
```

### Graphing Games Back

First we create a modified column to numerically represent 'Games Back'

```python
# Insert new column called 'int_GB' that represents GB as either negative (meaning behind) or 0.0 meaning in first or tied for first
def modify_gb(gb):
    if gb == 'Tied':
        gb = 0.0
    elif 'up' in gb:
        gb = 0.0 # float(gb.replace('up ', ''))
    else:
        gb = -float(gb)
    
    return gb
    
mets['int_GB'] = mets['GB'].dropna().apply(modify_gb)
braves['int_GB'] = braves['GB'].dropna().apply(modify_gb)
nationals['int_GB'] = nationals['GB'].dropna().apply(modify_gb)
phillies['int_GB'] = phillies['GB'].dropna().apply(modify_gb)
marlins['int_GB'] = marlins['GB'].dropna().apply(modify_gb)
```

#### What the Graph Shows
Graphing 'Games Back' gives an even clearer image of how the NL East was contended throughout the year. 

The Marlins fell to an early 8 Games Back, and never recovered. They ended the year at a season low of 40 Games Back.

The Mets, while contending early on, fell to ~13 Games Back mid-season. They showed signs of life in games 90-120, even surpassing the Phillies for 3rd place. However, the hole that was dug earlier in the season proved insurmountable, finsishing 11 games back.

The Phillies started the season hot and gained sole possesion of 1st place for the first third of the season.  Things would quickly turn, however, as the Phillies would spiral down to finish the season in 4th place, 16 games behind.  The Braves conversely started a bit rougher, however quickly covered ground in games 40-65 and overtook the Phillies for 1st place, which they would hold fairly comfortably for the remainder of the season.

The Nationals season was a roller coaster.  Dropping to a season low 10 games behind and 4th place in the first third of 2019, they went on to surpass both the Mets and Phillies and shot up to 2nd place by the final third of the season, which they would hold to the conclusion, finishing only 4 games behind Atlanta, good enough for the first wild card birth in the NL. 

```python
# Graph NL East Games Back Comparison #
plt.rcParams["figure.figsize"] = (8,6)

plt.plot(mets['int_GB'], label='Mets', c='darkorange')
plt.plot(braves['int_GB'], label='Braves', c='navy')
plt.plot(nationals['int_GB'], label='Nationals', c='gray')
plt.plot(phillies['int_GB'], label='Phillies', c='red')
plt.plot(marlins['int_GB'], label='Marlins', c='aqua')

plt.xticks(np.arange(0, len(braves.index), step=10))
plt.yticks(np.arange(0, -50, step=-2))
plt.xlabel('Game')
plt.ylabel('Games Back (in negative)')

plt.title("NL East Games Back Comparison ({})".format(year))
plt.legend(loc='lower left')
```

## Graphing Total Runs and Runs Allowed

First we create new columns for Total Runs, and Total Runs Allowed

```python
# Get total runs-to-date and total runs_allowed-to-date for each team
mets['Total_Runs'] = mets['R'].cumsum()
mets['Total_RA'] = mets['RA'].cumsum()

braves['Total_Runs'] = braves['R'].cumsum()
braves['Total_RA'] = braves['RA'].cumsum()

nationals['Total_Runs'] = nationals['R'].cumsum()
nationals['Total_RA'] = nationals['RA'].cumsum()

phillies['Total_Runs'] = phillies['R'].cumsum()
phillies['Total_RA'] = phillies['RA'].cumsum()

marlins['Total_Runs'] = marlins['R'].cumsum()
marlins['Total_RA'] = marlins['RA'].cumsum()
```

#### What the Graph Shows
Graphing Runs and Runs Allowed illustrates some of the causal factors in each teams divisional rank.

The Marlins were not scoring Runs at a rate similar to any of the other teams all season long, lagging starkly behind the competition. However, they competed in Runs Allowed with the others in division for the first 2/3 of 2019. By the end of the season, the Marlins had given up about 800 runs edging out the Phillies for worst in the division.

The Phillies had a more volatile season.  They scored at comparable and even higher rates to the rest of the division in the forst third of 2019, but ended up behind the Mets for 4th in scoring by the end of the year. Their pitching fared exremely well in the first third, allowing the least amount of runs, but ended up second worst by the end of the year challenging the Marlins for most runs allowed.

The Nationals and Braves were scoring at rates comparable to the rest of the division for the first third of 2019, then they broke off from the 3 others, battling it out for 1st and 2nd respectfully. In Runs Allowed the Braves were consistently about 3rd for the duration of the season while the Nationals fluctuated from the worst pitching staff through the first third of 2019, to the best in the division by the end.

The Mets offense performed fairly consistently in third place throughout 2019 finishing 3rd in runs scored, grouped closely with the Phillies but definitively behind the Braves and Nationals. Their pitching on the other hand was extremely volatile.  Immediately they were allowing significantly more runs than other team, then came back to comparable rates for games 35-70 as the other 4, then jumped back to the worst in the division for games 70-100, until ultimately preventing runs at a comparable and even better rate than the others finsishing 2019 as the second best pitching staff in the division.

```python
# Graph Runs and Runs Allowed Comparisons #
plt.rcParams["figure.figsize"] = (16,8)

plt.subplot(1,2,1)
plt.plot(mets['Total_Runs'], label='Mets', c='darkorange')
plt.plot(braves['Total_Runs'], label='Braves', c='navy')
plt.plot(nationals['Total_Runs'], label='Nationals', c='gray')
plt.plot(phillies['Total_Runs'], label='Phillies', c='red')
plt.plot(marlins['Total_Runs'], label='Marlins', c='aqua')

plt.xticks(np.arange(0, len(braves.index), step=10))
plt.xlabel('Game')
plt.ylabel('Runs')

plt.title("NL East Runs Comparison ({})".format(year))
plt.legend(loc='lower right')

plt.subplot(1,2,2)
plt.plot(mets['Total_RA'], label='Mets', c='darkorange')
plt.plot(braves['Total_RA'], label='Braves', c='navy')
plt.plot(nationals['Total_RA'], label='Nationals', c='gray')
plt.plot(phillies['Total_RA'], label='Phillies', c='red')
plt.plot(marlins['Total_RA'], label='Marlins', c='aqua')

plt.xticks(np.arange(0, len(braves.index), step=10))
plt.xlabel('Game')
plt.ylabel('Runs Allowed')

plt.title("NL East Runs Allowed Comparison ({})".format(year))
plt.legend(loc='lower right')
```

### Graphing Total Home Attendance
For fun, we take a look at each teams total home attendance throughout the season.

First we get each teams total home_attendance-to-date

```python
mets['Total_Home_Attendance'] = mets[mets['Home_Away'] == 'Home']['Attendance'].dropna().cumsum()
braves['Total_Home_Attendance'] = braves[braves['Home_Away'] == 'Home']['Attendance'].dropna().cumsum()
nationals['Total_Home_Attendance'] = nationals[nationals['Home_Away'] == 'Home']['Attendance'].dropna().cumsum()
phillies['Total_Home_Attendance'] = phillies[phillies['Home_Away'] == 'Home']['Attendance'].dropna().cumsum()
marlins['Total_Home_Attendance'] = marlins[marlins['Home_Away'] == 'Home']['Attendance'].dropna().cumsum()
```

#### What the Graph Shows
The Marlins saw dramatically less people come to the ballpark compared to the rest of the division, coming in last with ~750k attendees.

The Phillies come in first slightly ahead of the Braves. This may be caused by Citizen Bank Park's slightly larger capacity of 43,035 compared to Truist Park's capacity of 41,500.


The Mets beat out the Nationals by about 250k attendees. Given Citi Field's capacity (41,922) and National Park's Capacity (41,313), this seems to show a greater interest from Mets fans than Nationals fans, despite the Nationals having a more successful season.


```python
# Graph Total Home Attendance Comparison #
plt.rcParams["figure.figsize"] = (8,8)

# Create new Series for home attendance #
mets_attendance = mets['Total_Home_Attendance'].dropna()
mets_attendance.index = range(len(mets_attendance))

braves_attendance = braves['Total_Home_Attendance'].dropna()
braves_attendance.index = range(len(braves_attendance))

nationals_attendance = nationals['Total_Home_Attendance'].dropna()
nationals_attendance.index = range(len(nationals_attendance))

phillies_attendance = phillies['Total_Home_Attendance'].dropna()
phillies_attendance.index = range(len(phillies_attendance))

marlins_attendance = marlins['Total_Home_Attendance'].dropna()
marlins_attendance.index = range(len(marlins_attendance))

plt.plot(mets_attendance, label='Mets', c='darkorange')
plt.plot(braves_attendance, label='Braves', c='navy')
plt.plot(nationals_attendance, label='Nationals', c='gray')
plt.plot(phillies_attendance, label='Phillies', c='red')
plt.plot(marlins_attendance, label='Marlins', c='aqua')

plt.xlabel('Game')
plt.ylabel('Total Season Attendance')

plt.title("NL East Total Home Attendance Comparison ({})".format(year))
plt.legend(loc='lower right')
```
